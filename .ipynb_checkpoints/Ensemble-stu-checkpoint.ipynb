```python
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
X, y = make_moons(n_samples=100, noise=0.25, random_state=3) #noise 값 늘리면 noise가 커짐.따로 설정 가능.
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                    random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1], c= y)
```

```python
forest = RandomForestClassifier(n_estimators=5, random_state=2) #tree 5개.
forest.fit(X_train, y_train)
```

```python
forest.estimators_ #base classifier가 저장되어있음(5개)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title("Tree {}".format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)
    
mglearn.plots.plot_2d_separator(forest, X_train, fill=True, ax=axes[-1, -1],
                                alpha=.4)
axes[-1, -1].set_title("Random Forest")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
#각 tree가 상이한 모습, feature 수 =2 --> 1.42312 -> feature 1개 정해서 randomforest
#tree수 늘리면 부드러워짐. variance 감소.
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
```

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=0)
#stratify = target의 비율 그대로 split하기 위해.  ex) 양성데이터가 모두 train 데이터에 들어갈 수 있음. y데이터로 stratify!!
```

```python
training_accuracy = []
test_accuracy = []

n_settings = [1, 2, 5, 10, 20, 50, 100] #tree의 수를 세팅.
for n in n_settings:
    clf = RandomForestClassifier(n_estimators=n)
    clf.fit(X_train, y_train)
    
    y_train_hat = clf.predict(X_train)
    y_test_hat = clf.predict(X_test)
    
    training_accuracy.append(accuracy_score(y_train, y_train_hat))
    test_accuracy.append(accuracy_score(y_test, y_test_hat))
```

```python
pd.DataFrame({"n_estimators":n_settings, "training accuracy": training_accuracy, "test accuracy": test_accuracy})
#training accuracy는 n이 증가할수록 증가.
#test accuracy도 증가하는 경향을 보임. 안정적이게됨. 무조건 크면 클수록 좋은건 아님. 
```

### feature importance

```python
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)
```

```python
clf.feature_importances_ #attribute 생성됨. 변수 순서대로 출력됨.
```

```python
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(np.arange(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)
```

```python
plot_feature_importances_cancer(clf)
#worst parameter 가 제일 큼. -> worst parameter 가 크다고 양성일 확률 올라가는건 아님. 방향성 의미 XXXXXX
# logistic regression에서 계수가 양,음에는 방향이 있음.(예시)
```

## Adaboost

```python
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                    random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=5, random_state=42)
ada.fit(X_train, y_train)
```

```python
ada.estimators_
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), ada.estimators_)):
    ax.set_title("Tree {}".format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)
    
mglearn.plots.plot_2d_separator(ada, X_train, fill=True, ax=axes[-1, -1], alpha=.4)
axes[-1, -1].set_title("AdaBoost")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.show()

#stump. -> misclassed 에 weight 증가. -> 각각 tree의 confidence를 곱해서 선을 합침.
```

## Bagging (9 logistic regression models + voting)

```python
from sklearn.linear_model import LogisticRegression

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=0)
```

```python
n_estimators = 9

logregs = []

for i in range(n_estimators):
    np.random.seed(i)
    idx = np.random.choice(X_train.shape[0], X_train.shape[0], replace=True)
    X_train_base = X_train[idx,:]
    y_train_base = y_train[idx]
    logregs.append(LogisticRegression().fit(X_train_base,y_train_base))
```

```python

logregs
idx
```

```python
y_test_hats=[]
for i in range(n_estimators):
    y_test_hats.append(logregs[i].predict(X_test))
y_test_hats
```

```python
y_test_hats = np.stack(y_test_hats).T
# list 에서 np.array로 변경됨.
```

```python
y_test_hats
```

```python
y_test_hats.shape
```

```python
from scipy import stats
y_test_hat_voted = stats.mode(y_test_hats, axis=1)[0].squeeze()
#squeeze() ->  쓸데없는 차원 날리기. shape => (n,1) -> (n)
```

```python
stats.mode(y_test_hats, axis=1)#[0]
```

```python
for i in range(n_estimators):
    print(accuracy_score(y_test, y_test_hats[:,i]))
print("______________________________________")
print(accuracy_score(y_test,y_test_hat_voted))
```
